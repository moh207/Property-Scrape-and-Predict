## Use dictionary to unify same words with different spelling

```python
replacement_dict = {
    'apartment': ['شقة', 'شقه'],
    'services': ['خدمات','الخدمات'],
    'installments': ['تقسيط','اقساط','installment','اقساط'],
    'finished': [ 'متشطب'],
    'compound': 'كمبوند',
    'furnished':['مفروش'],
    'fifth': ['خامس'],
    'settlement':['تجمع'],
    'mall':['مول'],
    'club':['نادي'],
    'located':['يقع'],
    'near':'القرب',
    'lux':'لوكس',
    'park':'بارك',
    'full':'كامل',
    'swimming':'سباحة',
    'minutes':'دقائق',
    'airport':'مطار',
    'zayed':'زايد',
    
    

    # Add more key-value pairs as needed
}

# Specify the column in which you want to replace words
column_to_replace = 'description_text'


def generate_regex(word):
    return fr'\b(?:\w{{0,2}}{re.escape(word)}\w{{0,2}})\b'


def replace_words(text):
    for key, values in replacement_dict.items():
        if isinstance(values, str):
            regex = generate_regex(values)
            text = re.sub(regex, key, text)
        else:
            for value in values:
                regex = generate_regex(value)
                text = re.sub(regex, key, text)
    return text


# Apply the replacement function to the specified column
df['new_txt'] = df[column_to_replace].str[len('Description'):].apply(replace_words)
```

```python
def add_space_between_words(text):
    # Define a regular expression pattern to identify boundaries
    pattern = re.compile(r'([^\d\s]+|\d+)')

    # Use the pattern to split the text and insert spaces
    result = ' '.join(pattern.findall(text))

    return result
```

```python
from langid import classify
from translate import Translator

def translate_mixed_language(text):
    # Detect the language of the entire sentence
    lang, _ = classify(text)

    # Translate the sentence if it's in Arabic
    if lang == 'ar':
        translator = Translator(to_lang="en", from_lang="ar")
        translated_text = translator.translate(text)
        return translated_text
    else:
        # If the sentence is in English or another language, keep it as it is
        return text

# Example usage
mixed_text = "مرحبا، كيف حالك يل محمد؟ Hello, how are you?"
translated_text = translate_mixed_language(mixed_text)

print(f"Original Text: {mixed_text}")
print(f"Translated Text: {translated_text}")
```

```python
import lightgbm as lgb
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.utils.validation import check_X_y, check_array, check_is_fitted

class LGBWrapper(BaseEstimator, RegressorMixin):
    def __init__(self, model=None):
        self.model = model
        self.is_fitted_ = True  # Flag to track whether the model is fitted

    def fit(self, X, y=None):
        check_is_fitted(self)  # Check if the model is already fitted
        X, y = check_X_y(X, y)
        lgb_train = lgb.Dataset(X, y)
        self.model = lgb.train(self.model.params, lgb_train, num_boost_round=self.model.best_iteration)
        self.is_fitted_ = True  # Set the fitted flag to True
        return self

    def predict(self, X):
        check_is_fitted(self)  # Check if the model is fitted before predicting
        #X = check_array(X)
        return self.model.predict(X)


#lgb_model = lgb.train(params, lgb.Dataset(X, y), num_boost_round=100)

# Create a LightGBM wrapper with the trained model
lgb_wrapper = LGBWrapper(model=lgb_model)
```
